```python
import numpy as np
import matplotlib.pyplot as plt


categories = ['Obedience', 'Security', ' To be Pleasing']
section1 = [5, 7, 8]  # Section 1 values for each category
section2 = [3, 5, 6]  # Section 2 values for each category
section3 = [2, 3, 4]  # Section 3 values for each category

# Number of categories
x = np.arange(len(categories))

# Plot stacked bars
plt.bar(x, section1, label='Primed', color='midnightblue')
plt.bar(x, section2, bottom=section1, label='Neutral', color='blue')
plt.bar(x, section3, bottom=np.array(section1) + np.array(section2), label='Negatively Primed', color='lightblue')

# Customizing the plot
plt.xlabel('Categories')
plt.ylabel('% of Promts returning Malware')
plt.title('Stacked Bar Chart Example')
plt.xticks(x, categories)  # Replace x-axis ticks with category names
plt.legend()  # Add legend
plt.grid(axis='y')  # Optional: Add gridlines to y-axis

# Show the plot
plt.tight_layout()
plt.show()
```
